```python
data.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
ethnic_group_data = data['EthnicGroup'].mode()[0]
```

```python
data['EthnicGroup'].fillna(ethnic_group_data,inplace=True)
```

```python
data
```

```python
ParentEduc_data = data['ParentEduc'].mode()[0]
```

```python
data['ParentEduc'].fillna(ParentEduc_data, inplace=True)
```

```python
TestPrep_data = data['TestPrep'].mode()[0]

data['TestPrep'].fillna(TestPrep_data,inplace=True)
```

```python
data.isnull().sum()
```

```python
data['ParentMaritalStatus'].fillna('single',inplace=True)
```

```python
PracticeSport_data = data['PracticeSport'].mode()[0]

data['PracticeSport'].fillna(PracticeSport_data,inplace=True)
```

```python
data['IsFirstChild'].value_counts()
```

```python
data['IsFirstChild']=data['IsFirstChild'].map({'yes': 0, 'no' :1})
```

```python
data['IsFirstChild'].fillna(np.random.randint(0,1),inplace=True)
```

```python
data['NrSiblings'].fillna(data['NrSiblings'].mean(),inplace=True)
```

```python
data['TransportMeans'].value_counts()
```

```python
data['TransportMeans'].map({'school_bus' : 0, 'private' :1})

data['TransportMeans'].fillna(np.random.randint(0,1),inplace=True)
```

```python
data['WklyStudyHours'].value_counts()
```

```python
data['WklyStudyHours'].map({'5 - 10' : 0, '< 5' : 1, '> 10' : 2})

data['WklyStudyHours'].fillna(np.random.randint(0,2) , inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Overall_Score'] = data['MathScore'] + data['ReadingScore'] + data['WritingScore']
```

```python
data
```

```python
class_group = sns.countplot(data , x = 'EthnicGroup' , hue = 'Gender')
plt.title('Classification of Ethnic Group on Gender')
plt.show()
```

```python
test_prep_aggregate = data.groupby('TestPrep')['Overall_Score'].mean()
plt.figure(figsize=(5,5))
test_prep_aggregate.plot(kind='bar')
plt.title('Impact of Test Preparation')
plt.xlabel('TestPrep')
plt.ylabel('Overall Score')
plt.xticks(rotation=0)
plt.show()
```

```python
gender_aggregate = data.groupby('Gender')['Overall_Score'].sum()
plt.figure(figsize=(5,5))
plt.pie(gender_aggregate, labels=gender_aggregate.index, autopct='%1.1f%%', startangle=50)
plt.title('Overall Score Distribution by Gender')
plt.show()
```

```python
weekly_study_aggregated = data.groupby('WklyStudyHours')['Overall_Score'].mean()
plt.figure(figsize=(5,5))
weekly_study_aggregated.plot(kind='bar')
plt.title('Impact of Weekly Study Preparation')
plt.xlabel('Weekly Study Hours')
plt.ylabel('Overall Score')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.hist(data['Overall_Score'], bins=10, edgecolor='black')
plt.title('Distribution of Overall Scores')
plt.xlabel('Overall Score')
plt.ylabel('Frequency')
plt.show()
```
